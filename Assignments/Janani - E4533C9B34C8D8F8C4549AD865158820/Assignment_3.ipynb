```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('/content/House Price India.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['Price'])
plt.show()
```

```python
#Bi-Variate Analysis:
# Example: Scatter plot for two numerical features
sns.scatterplot(x='Built Year', y='Price', data=df)
plt.show()
```

```python
#Multi-Variate Analysis:
sns.pairplot(df[['Price', 'number of bedrooms', 'number of bathrooms']])
plt.show()
```

```python
correlation_matrix = df[['Price', 'number of bedrooms', 'number of bathrooms']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
# Example: Summary statistics for the entire dataset
summary = df.describe()
print(summary)
```

```python
# Example: Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```
